```python
ice_cream_rating = 2
sleep_rating = 4
```

```python
first_name = 'Фёдор'
last_name = 'Рочев '
my_name = last_name + first_name
```

```python
happiness_rating = (ice_cream_rating + sleep_rating) / 2
```

```python
print(type(first_name), type(ice_cream_rating), type(happiness_rating))
```

Все типы данных имели смысл, типов данных которые отличались от моих ожиданий нет.

```python
print('Меня зовут', first_name, ', и я ставлю мороженому', ice_cream_rating, 'баллов из 10!')
print('Я', my_name,', и мой рейтинг удовольствия от сна — ', sleep_rating, ' из 10!')
print('Исходя из вышеперечисленных факторов, мой рейтинг счастья составляет', happiness_rating, 'из 10, или ', happiness_rating/10*100,'%!')
```
