```python
import pandas as pd
```

```python
df=pd.read_csv('data/2315676.csv', na_values=-9999, sep='\s+', skiprows=[1])
```

```python
df.head()
```

```python
df_1984=df.loc[((df['DATE']>=19840101) & (df['DATE']<=19841231))]
```

```python
df_1984
```

```python
df_1984['TAVG_NEW']=df_1984['TAVG'].fillna((df['TMAX']+df['TMIN'])/2)
```

```python
df_1984['TAVG_NEW']=(df_1984['TAVG_NEW']-32)/1.8
```

```python
temps_winter=df_1984.loc[(((df_1984['DATE']>=19840101) & (df_1984['DATE']<=19840229)) | ((df_1984['DATE']>=19841201) & (df_1984['DATE']<=19841231)))]
temps_winter['DATE']=pd.to_datetime(temps_winter['DATE'], format='%Y%m%d')
temps_winter
```

```python
temps_spring=df_1984.loc[((df_1984['DATE']>=19840301) & (df_1984['DATE']<=19840531))]
temps_spring['DATE']=pd.to_datetime(temps_spring['DATE'], format='%Y%m%d')
temps_spring
```

```python
temps_summer=df_1984.loc[((df_1984['DATE']>=19840601) & (df_1984['DATE']<=19840831))]
temps_summer['DATE']=pd.to_datetime(temps_summer['DATE'], format='%Y%m%d')
temps_summer
```

```python
temps_autumn=df_1984.loc[((df_1984['DATE']>=19840901) & (df_1984['DATE']<=19841130))]
temps_autumn['DATE']=pd.to_datetime(temps_autumn['DATE'], format='%Y%m%d')
temps_autumn
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize = (8,8))
plt.plot(temps_winter['DATE'],temps_winter['TAVG_NEW'])
plt.title('Winter')
```

```python
fig, ax = plt.subplots(figsize = (8,8))
plt.plot(temps_spring['DATE'],temps_spring['TAVG_NEW'], color='orange')
plt.title('Spring')
```

```python
fig, ax = plt.subplots(figsize = (8,8))
plt.plot(temps_summer['DATE'],temps_summer['TAVG_NEW'], color='green')
plt.title('Summer')
```

```python
fig, ax = plt.subplots(figsize = (8,8))
plt.plot(temps_autumn['DATE'],temps_autumn['TAVG_NEW'], color='red')
plt.title('Autumn')
```
